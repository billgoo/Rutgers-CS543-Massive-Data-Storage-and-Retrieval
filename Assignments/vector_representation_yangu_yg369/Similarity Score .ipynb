```python
import math
import numpy as np
# the code is too easy and naive so I don't think it will need much comment
```

```python
# load glove model
def load_glove_model(gloveFile):
    print("Loading Model . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    f = open(gloveFile,'r', encoding="utf8")
    model = {}
    counter = 0
    for line in f:
        counter+=1
        try:
            splitLine = line.split()
            word = splitLine[0]
            embedding = np.array([float(val) for val in splitLine[1:]])
            model[word] = embedding
        except Exception as e:
            continue
#             print('error = {0}, input = {1}'.format(e,line))
        if counter % 100000 == 0 :
            print("Done.",len(model)," words loaded . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    print("Done.",len(model)," words loaded!")
    return model
```

```python
# compute cosine
def compute_cosine(vector1, vector2):
    numerator = 0
    norm1 = 0
    norm2 = 0
    for i in range(len(vector1)):
        numerator += vector1[i] * vector2[i]
        norm1 += pow(vector1[i], 2)
        norm2 += pow(vector2[i], 2)

    denominator = math.sqrt(norm1) * math.sqrt(norm2)
    try:
        similarity = float(numerator) / denominator
    except ZeroDivisionError:
        similarity = 0.0

    return similarity
```

```python
def get_sentence_vector(model, sentence):
    sentence = sentence.lower()
    
    # tokeniz
    tokenize = sentence.split(" ")
    words = []

    for token in tokenize:
        if token in model.keys():
            words.append(token)

    sentence_vector = [0 for _ in range(300)]
    for word in words:
        for i in range(len(sentence_vector)):
            sentence_vector[i] += model[word][i]

    sentence_vector = [i / len(words) for i in sentence_vector]
    return sentence_vector
```
